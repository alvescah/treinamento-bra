### 1. Some todos os números de 1 até 100

```python
count = 0
num = 0

while count <= 100:
  num = num + count
  count += 1
  
print(num)
```

### 2. Some todos os números pares de 1 até 100

```python
count = 0
num = 0

while count <= 100:
  if count % 2 == 0:
    num = num + count
  count += 1
  
print(num)
```

### 3. Some todos os números ímpares de 1 até 100

```python
count = 0
num = 0

while count <= 100:
  if count % 2 != 0:
    num = num + count
  count += 1
  
print(num)
```

### 4. Dada a lista abaixo, some todos os elementos inteiros positivos

```python
lista = [-2, 2, 'celular', 10, -99, False]
soma = 0

for num in lista:
  if type(num) == int:
    if num > 0:
      soma += num

print(soma)
    
```
